# **Group and aggregate data using groupby**

```python
data = {
    "Department": ["HR", "Finance", "HR", "IT", "Finance", "IT"],
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "Salary": [50000, 60000, 55000, 70000, 65000, 72000],
    "Experience": [2, 5, 3, 8, 6, 7],
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

# **Grouping Data**

## **sum() calculates the total of groupd values**

```python
df.groupby("Department")["Salary"].sum()
```

# **mean()**

**mode()**

**median()**

**of grouped values.**

```python
print(df.groupby("Department")["Salary"].mean())
```

```python
df.groupby("Department")["Salary"].median()
```

**min()/max() finds the minimum or maximum value in each group.**

```python
df.groupby("Department")["Salary"].min()
```

```python
df.groupby("Department")["Salary"].max()
```

**count()**

**counts the number of non-null values in each group.**

```python
df.groupby("Department")["Salary"].count()
```

**std()/var()**

**calculates the standard deviation or variance of grouped values.**

```python
df.groupby("Department")["Salary"].std()
```

```python
df.groupby("Department")["Salary"].var()
```

**first()/last()**

**Returns the first or last non-null value in each group.**

```python
df.groupby("Department")["Salary"].last()
```

```python
df.groupby("Department")["Salary"].first()
```

**nunique()**

**Counts the number of unique values in each group.**

```python
df.groupby("Department")["Employee"].nunique()
```

**prod()**

**Computes the product of grouped values.**

```python
df.groupby("Department")["Salary"].prod()
```

**Using agg() for Multiple Aggregations**

**The agg method allows applying multiple aggregation functions simultaneously,including custom functions.**

**Single Column,Multiple Functions**

```python
df.groupby("Department")["Salary"].agg(["mean", "max", "min"])
```

**Multiple Columns,Different Functions**

```python
df.groupby("Department").agg({
    "Salary": ["mean", "max"],
    "Experience": ["sum", "min"]
})
```

**Custom**

```python
df.groupby("Department")["Salary"].agg(lambda x: x.max() - x.min())
```
