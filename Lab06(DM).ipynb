```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
#example data
data={'category':['A','B','C','A','B']}
df=pd.DataFrame(data)
df
```

```python
#label encoding
label_encoder=LabelEncoder()
df['Label_Encoder']=label_encoder.fit_transform(df['category'])
```

```python
#one hot encoding
one_hot_encoder=pd.get_dummies(df['category'],prefix='category')
df=pd.concat([df,one_hot_encoder],axis=1)
df
```

```python
from datetime import datetime
```

```python
#Example data
data={'Name': ['Harini','Ravi'],'Date_of_Birth':['2006-01-19','1997-11-20']}
df=pd.DataFrame(data)
df
```

```python
#convert Date_of_Birth to datetime
df['Date_of_Birth']=pd.to_datetime(df['Date_of_Birth'])
```

```python
#calculate Age
current_year=datetime.now().year
df['Age']=current_year-df['Date_of_Birth'].dt.year
df
```

```python
#Example data
data={'category': ['A','A','B','B','C'],'Value':[10,20,10,30,40]}
df=pd.DataFrame(data)
```
